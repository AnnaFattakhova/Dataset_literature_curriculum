**2. Соотношение поэзия/проза/inapp в динамике.**

```python
res_df_2 = pd.DataFrame(columns=['source_year'] + categories_2)

for i in dictionary_programs:
    count_dict = {category: 0 for category in categories_2}
    for index, row in df.iterrows():
        if row['source_year'] == i and row['genre'] in count_dict:
            count_dict[row['genre']] += 1

    res_df_2.loc[len(res_df_2)] = [i] + [count_dict[category] for category in categories_2]
res_df_2.set_index('source_year').T
```

**7. Индивидуальные траектории авторов (таблица автор/годы с отмеченной на их пересечении представленностью в программах).**

| |1974/75|1975/76|
|--------|-------|-------|
|автор 1 |     1 |      1|
|автор 2 |      1|      0|
|автор 3 |      0|      1|

pd.crosstab() — это функция Pandas, которая создает кросс-таблицу (сводную таблицу) по двум (или более) столбцам DataFrame, подсчитывая частоту встречаемости значений.

*table = pd.crosstab(df['Author'], df['Year'])*

По умолчанию crosstab() считает количество, но можно преобразовать в бинарный формат:

*table = pd.crosstab(df['Author'], df['Year']).astype(bool).astype(int)*


```python
# Это с количеством авторов в программах
res_df_7 = pd.crosstab(df['author'], df['source_year'])
res_df_7
```

```python
res_df_7 = pd.crosstab(df['author'], df['source_year']).astype(bool).astype(int)
res_df_7
```
